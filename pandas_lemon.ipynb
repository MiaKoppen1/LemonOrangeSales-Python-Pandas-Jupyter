```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_data = pd.read_csv('Lemonade2016-2.csv')
lemon_orange_sales = pd.DataFrame(my_data)
pd.isnull(lemon_orange_sales) #will say true for all values that are null
```

```python
lemon_orange_sales.fillna("x") #replaces all null values with "x"
```

```python
price_mean = lemon_orange_sales['Price'].mean()
lemon_orange_sales.fillna(price_mean)
```

```python
lemon_orange_sales.duplicated(subset=None, keep=False) #shows duplicates as "true"
```

```python
lemon_orange_sales.drop_duplicates() #drops all duplicated from the Data Frame
```

```python
lemon_orange_sales.head(5)
```

```python

lemon_orange_sales.loc[:, ['Price', 'Orange', 'Lemon']]
```

```python
lemon_orange_sales['Sales'] = lemon_orange_sales['Orange'] + lemon_orange_sales['Lemon']
plt.scatter(lemon_orange_sales['Leaflets'], lemon_orange_sales['Sales'])

```

```python
lemon_orange_sales['Revenue'] = lemon_orange_sales['Sales'] * lemon_orange_sales['Price']
plt.scatter(lemon_orange_sales['Revenue'], lemon_orange_sales['Sales'])
```

```python
plt.hist(lemon_orange_sales['Revenue'])
```

```python
lemon_orange_sales.boxplot(column='Sales', by='Location')
plt.ylabel('Sales')
plt.title('Total Sales by Location')
plt.suptitle('')  # Suppress default title
plt.show()
```

```python
selected_columns = ['Location', 'Revenue', 'Temperature', 'Leaflets']
lemon_orange_selected = lemon_orange_sales[selected_columns]

# Create a pairplot using seaborn
seaborn_pair = sns.pairplot(lemon_orange_selected)
```
